```python
# This script aims to summaries the manual annotation for video score of JAK1i project
import numpy as np
import pandas as pd

csvFileName = 'C:\\Python27\\Scripts\\ipython_notebook\\JD_Cam2_20150514Cages1_2_features_predicted.csv'

df = pd.read_csv(csvFileName)

df.head(5)
```

```python
JD = df['JD_score_group'].values
SMO = df['SMO_score_group'].values
COM = df['predict'].astype(int).values
print type(JD)
```

```python
import matplotlib.pyplot as plt
ind1 = np.arange(140*15,140*15+60*15)
ind2 = np.arange(600*15,600*15+60*15)
ind3 = np.arange(2680*15,2680*15+60*15)

plt.subplot(311)
plt.plot(ind1/15.0/60.0,JD[ind1],'b',linewidth = 1.5,label = 'JD')
plt.plot(ind1/15.0/60.0,SMO[ind1],'k',linewidth = 1.5,label = 'SMO')
plt.plot(ind1/15.0/60.0,COM[ind1],'r',linewidth = 2.5,label = 'COM')
plt.xlim(ind1[1]/900.0,ind1[-1]/900.0)
plt.ylim(-0.5,2.5)
plt.grid()
plt.ylabel('Seizur Stages',fontsize = 20)
plt.xlabel('Time(minutes)',fontsize = 16)
plt.yticks((0, 1, 2), ('0','1-3','4-6'))
plt.legend()
plt.title('Case 1')


plt.subplot(312)
plt.plot(ind2/15.0/60.0,JD[ind2],'b',linewidth = 1.5,label = 'JD')
plt.plot(ind2/15.0/60.0,SMO[ind2],'k',linewidth = 1.5,label = 'SMO')
plt.plot(ind2/15.0/60.0,COM[ind2],'r',linewidth = 2.5,label = 'COM')
plt.xlim(ind2[1]/900.0,ind2[-1]/900.0)
plt.ylim(-0.5,2.5)
plt.grid()
plt.ylabel('Seizur Stages',fontsize = 20)
plt.xlabel('Time(minutes)',fontsize = 16)
plt.yticks((0, 1, 2), ('0','1-3','4-6'))
plt.legend()
plt.title('Case 2')

plt.subplot(313)
plt.plot(ind3/15.0/60.0,JD[ind3],'b',linewidth = 1.5,label = 'JD')
plt.plot(ind3/15.0/60.0,SMO[ind3],'k',linewidth = 1.5,label = 'SMO')
plt.plot(ind3/15.0/60.0,COM[ind3],'r',linewidth = 2.5,label = 'COM')
plt.xlim(ind3[1]/900.0,ind3[-1]/900.0)
plt.ylim(-0.5,2.5)
plt.grid()
plt.ylabel('Seizur Stages',fontsize = 20)
plt.xlabel('Time(minutes)',fontsize = 16)
plt.yticks((0, 1, 2), ('0','1-3','4-6'))
plt.legend()
plt.title('Case 3')

plt.tight_layout()
plt.show()
```
