```python
nrow,ncol = df.shape

activitiesNames = ['0','1-3','4-6']


ts = np.linspace(0,(nrow-1)/900.0,nrow)
print len(ts), len(df['predict'])

#140,602,2678

ind = range(140*15,140*15+15*60)
            
plt.subplot(311)
plt.plot(ts[ind],df['predict'][ind],'r',linewidth = 3.0,label = "COM")
plt.plot(ts[ind],df['JD_score_group'][ind],'b',linewidth = 1.0,label = "JD")
plt.plot(ts[ind],df['SMO_score_group'][ind],'k',linewidth = 1.0,label = "SMO")
plt.grid()
plt.ylim(-0.5,2.5)
plt.xlim(ts[ind[0]],ts[ind[-1]])
plt.xlabel('Time (minutes)',fontsize = 20)
plt.ylabel('Seizure stages',fontsize = 20)
tick_marks = np.arange(len(activitiesNames))
plt.yticks(tick_marks, activitiesNames,fontsize=15)
plt.legend(bbox_to_anchor=(.90, 0.90), loc=2, borderaxespad=0.1)
plt.title('Case 1')

ind = range(602*15,602*15+15*60)
            
plt.subplot(312)
plt.plot(ts[ind],df['predict'][ind],'r',linewidth = 3.0,label = "COM")
plt.plot(ts[ind],df['JD_score_group'][ind],'b',linewidth = 1.0,label = "JD")
plt.plot(ts[ind],df['SMO_score_group'][ind],'k',linewidth = 1.0,label = "SMO")
plt.grid()
plt.ylim(-0.5,2.5)
plt.xlim(ts[ind[0]],ts[ind[-1]])
plt.xlabel('Time (minutes)',fontsize = 20)
plt.ylabel('Seizure stages',fontsize = 20)
tick_marks = np.arange(len(activitiesNames))
plt.yticks(tick_marks, activitiesNames,fontsize=15)
plt.legend(bbox_to_anchor=(.90, 0.90), loc=2, borderaxespad=0.1)
plt.title('Case 2')

ind = range(2678*15,2678*15+15*60)
            
plt.subplot(313)
plt.plot(ts[ind],df['predict'][ind],'r',linewidth = 3.0,label = "COM")
plt.plot(ts[ind],df['JD_score_group'][ind],'b',linewidth = 1.0,label = "JD")
plt.plot(ts[ind],df['SMO_score_group'][ind],'k',linewidth = 1.0,label = "SMO")
plt.grid()
plt.ylim(-0.5,2.5)
plt.xlim(ts[ind[0]],ts[ind[-1]])
plt.xlabel('Time (minutes)',fontsize = 20)
plt.ylabel('Seizure stages',fontsize = 20)
tick_marks = np.arange(len(activitiesNames))
plt.yticks(tick_marks, activitiesNames,fontsize=15)
plt.legend(bbox_to_anchor=(.90, 0.90), loc=2, borderaxespad=0.1)
plt.title('Case 3')

plt.show()
#df.to_csv(featureFileFullNameSave)
print ts[-1]
```
